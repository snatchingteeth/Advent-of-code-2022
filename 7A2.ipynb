```python
import csv
data = open("input(day7).csv")
csv_data = csv.reader(data)
data_lines = list(csv_data)
```

```python
directories = {}
```

```python
#filling the dictionary
for line_num in range(0,len(data_lines)):
    if data_lines[line_num] ==['$ ls']:
        current_directory= data_lines[line_num-1][0][5:]
        directories[current_directory]=[]
    elif data_lines[line_num][0][0] =="$":
        pass
    else: 
        directories[current_directory].append(data_lines[line_num][0])
```

```python
directories
```

```python
#file totals
totals = {}
num_dirs_in_each_dir = {}
for directory in directories:
    totals[directory] = 0
    num_dirs_in_each_dir[directory] = 0
    for item in directories[directory]:
        if item[0] != "d":
            space = item.index(" ")
            totals[directory] += int(item[:space])
        else:
           num_dirs_in_each_dir[directory] +=1
```

```python
totals
```

```python
for x in range(0, max(num_dirs_in_each_dir.values())+1):
    for directory in directories:
        if num_dirs_in_each_dir[directory] ==x:
           for item in directories[directory]:
                if item[0] == "d":
                    totals[directory] += totals[item[4:]] 
```

```python
final = 0
```

```python
for dictionary in totals:
    if totals[dictionary] <= 100000:
        final += totals[dictionary]
```
